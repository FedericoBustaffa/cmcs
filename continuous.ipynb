# Continuous Systems


```python
import numpy as np
import matplotlib.pyplot as plt

def d_birth(n0, rate, t):
    if t == 0:
        return n0
    return n0 * np.pow(rate, t-1)

def c_birth(n0, rate, t):
    return n0 * np.exp(rate * t)

time_span = 10
N_0 = 5
d_rate = 2
d_time = [i for i in range(time_span + 1)]
discrete = [d_birth(N_0, d_rate, t) for t in d_time]

c_rate = np.log(d_rate)
c_time = np.linspace(0, time_span, 50)
continuous = [c_birth(N_0, c_rate, t) for t in c_time]

plt.figure(figsize=(8, 6))
plt.plot(d_time, discrete, label="discrete")
plt.plot(c_time, continuous, label="continuous")
plt.legend()
plt.grid()
plt.show()
```
